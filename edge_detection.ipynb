```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image_path =r'kitty.jpg'
img=cv2.imread(image_path, cv2.IMREAD_COLOR)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges_x=cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
edges_y=cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
```

```python
edges=cv2.magnitude(edges_x, edges_y)
plt.figure(figsize=(12, 6))
plt.subplot(121), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.subplot(122), plt.imshow(edges, cmap='gray'), plt.title('Edge-detected image')
plt.show()
```
